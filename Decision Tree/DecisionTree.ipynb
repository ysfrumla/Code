```python
import sklearn.tree as tree
import sklearn.metrics as met 
import pandas as pd 
import sklearn.model_selection as ms 


df1 = pd.read_csv(r"E:\File Program\Semester II\Jupyter Notebook\Data Latihan\data\New folder (3)\decisiontree_ch6.csv")
encoding = {
    "mesin": {"bensin": 0, "diesel": 1},
    "penggerak": {"depan": 0, "belakang": 1}
}

df1.replace(encoding, inplace = True)

X = df1.drop(['ID', 'label'], axis=1)
y = df1['label']

X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size = 0.2)
modell = tree.DecisionTreeClassifier(criterion = 'gini', max_depth = 5)
modell.fit(X_train, y_train)
y_prediksi = modell.predict(X_test)
print(y_prediksi)

a = met.accuracy_score(y_test, y_prediksi)
print(str(a * 100) + '%')
```

```python
import pydotplus as pp

labels = ['mesin', 'bangku', 'penggerak']
dot_data = tree.export_graphviz(modell, out_file = None, feature_names = labels, filled = True, rounded = True)
graph = pp.graph_from_dot_data(dot_data)
graph.write_png('decisiontree.png')
#print(graph)
```
